## Appreciation for Wildlife

```python
# df_appreciation_wl = new_df[['video_url', 'comment', 'appreciation_wildlife']]
# df_appreciation_wl
```

```python
# conditions = [
#     df_appreciation_wl['appreciation_wildlife'] >= 1,
#     df_appreciation_wl['appreciation_wildlife'] < 1
# ]
# choices = [1, 0]

# df_appreciation_wl['appreciation_wildlife'] = np.select(conditions, choices, default=df_appreciation_wl['appreciation_wildlife'])
# df_appreciation_wl
```

```python
# df_appreciation_wl_majority = df_appreciation_wl.groupby(['video_url', 'comment'])['appreciation_wildlife'].agg(lambda x: x.mode().iloc[0]).reset_index()
# df_appreciation_wl_majority
```

```python
df_appreciation_wl = new_df[['video_url', 'comment', 'appreciation_wildlife']]
df_appreciation_wl = df_appreciation_wl[['video_url','comment','appreciation_wildlife']].replace('',0)
df_appreciation_wl = df_appreciation_wl[df_appreciation_wl['video_url'] != 0]

conditions = [
    df_appreciation_wl['appreciation_wildlife'] >= 1,
    df_appreciation_wl['appreciation_wildlife'] < 1
]
choices = [1, 0]

df_appreciation_wl['appreciation_wildlife'] = np.select(conditions, choices, default=df_appreciation_wl['appreciation_wildlife'])
df_appreciation_wl_majority = df_appreciation_wl.groupby(['video_url', 'comment'])['appreciation_wildlife'].agg(lambda x: x.mode().iloc[0]).reset_index()
df_appreciation_wl_majority
```

```python
df_appreciation_wl_majority.groupby('appreciation_wildlife').count().reset_index()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000) 

tfidf_matrix = vectorizer.fit_transform(df_appreciation_wl_majority['comment'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

labels = df_appreciation_wl_majority[['appreciation_wildlife']]
# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, labels, test_size=0.2, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate predictions
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
f1 = f1_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
print("f1_score:", f1)
print("precision:", precision)
print("recall:", recall)
print("accuracy:", accuracy)
```

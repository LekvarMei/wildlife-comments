## fleiss_kappa & Krippendorf's alpha

```python
df_appreciation_wl = new_df[['video_url', 'comment', 'appreciation_wildlife']]
df_appreciation_wl = df_appreciation_wl[['video_url','comment','appreciation_wildlife']].replace('',0)

conditions = [
    df_appreciation_wl['appreciation_wildlife'] > 0,
    df_appreciation_wl['appreciation_wildlife'] <= 0
]
choices = [1, 0]

df_appreciation_wl['appreciation_wildlife'] = np.select(conditions, choices, default=df_appreciation_wl['appreciation_wildlife'])
df_appreciation_wl
```

```python
df_appreciation_wl['Appreciation Count'] = df_appreciation_wl.groupby(['video_url', 'comment']).cumcount() + 1

# Group by comment and Appreciation Count, then unstack
df_pivot = df_appreciation_wl.groupby(['video_url','comment', 'Appreciation Count'])['appreciation_wildlife'].aggregate('sum').unstack()

# Fill NaN values with 0, since they indicate no ratings were given
df_pivot = df_pivot.fillna(0)

# Convert to integers if not already
df_pivot = df_pivot.astype(int)

# Rename columns to match the expected format
df_pivot.columns = [f'Appreciation {i}' for i in range(1, len(df_pivot.columns) + 1)]

# Reset index to make 'comment' a column again if needed
df_pivot.reset_index(inplace=True)

df_pivot
```

```python



from statsmodels.stats.inter_rater import fleiss_kappa

# Drop the 'video_url' column as it's not needed for the calculation
df_for_kappa = df_pivot.drop(columns=['video_url'])

df_for_kappa = df_for_kappa.set_index('comment')
df_for_kappa = df_for_kappa.apply(pd.value_counts, axis=1).fillna(0)

# Now calculate Fleiss' Kappa
kappa = fleiss_kappa(df_for_kappa)

print('Fleiss Kappa:', kappa)

```

```python
import krippendorff

# Assuming you've already loaded your DataFrame as df_pivot and dropped the 'video_url' column
df_for_alpha = df_pivot.drop(columns=['video_url'])

# The DataFrame for Krippendorff's Alpha should be in the format of:
# rows are units, columns are raters, and cells contain the ratings
df_for_alpha = df_for_alpha.set_index('comment')

# Now we'll prepare the data in the correct format for krippendorff.alpha
# Convert the DataFrame to a format where each row is an item and each column is a rater's scores
alpha_data = df_for_alpha.T.to_numpy()

# Calculate Krippendorff's Alpha for nominal metric
alpha = krippendorff.alpha(alpha_data, level_of_measurement='nominal')

print('Krippendorff\'s Alpha:', alpha)
```
